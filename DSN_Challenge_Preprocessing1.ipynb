```python
# Here we change the customer Id into integer variable
train_data['Customer Id'] = train_data['Customer Id'].apply(lambda s : int(s.strip("H")))
test_data['Customer Id'] = test_data['Customer Id'].apply(lambda s : int(s.strip("H")))
```

```python
# Here we change the Building_Type variable into a categorical varible
train_data = train_data.astype({'Building_Type': str})
test_data = test_data.astype({'Building_Type': str})
```

```python
# Getting list of categorical variables
s = (train_data.dtypes == 'object') # gives a series with True and False according to the column type
cat_cols = list(s[s].index) # gives the cat var names
print("Categorical variables:")
print(cat_cols)
```

```python
#instantiate both packages to use
encoder = OrdinalEncoder()
imputer = IterativeImputer(ExtraTreesRegressor())
# create a list of categorical columns to iterate over
cat_cols = ['Building_Painted', 'Building_Fenced', 'Garden', 'Settlement', 'Building_Type', 'NumberOfWindows', 'Geo_Code']

def encode(data):
    '''function to encode non-null data and replace it in the original data'''
    #retains only non-null values
    nonulls = np.array(data.dropna())
    #reshapes the data for encoding
    train_data_reshape = nonulls.reshape(-1,1)
    #encode data
    train_data_ordinal = encoder.fit_transform(train_data_reshape)
    #Assign back encoded values to non-null values
    data.loc[data.notnull()] = np.squeeze(train_data_ordinal)
    return data

#create a for loop to iterate through each column in the data
for columns in cat_cols:
    encode(train_data[columns])
```
